```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
#ratings.head()
rate = [5.0]
result = ratings [
    ratings['rating'].isin(rate)]
#result.sort_values('movieId')
result.groupby('movieId').count().head(1)
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd
power = pd.read_csv('power.csv')
baltic_countries = ['Latvia','Estonia','Lithuania']
category = [4, 12, 21]
years = range(2005, 2011)
result = power [
    power['country'].isin(baltic_countries) &
    power['category'].isin(category) &
    power['year'].isin(years) &
    power['quantity'] > 0
]
#result[['country','quantity']].groupby('country').sum()
result[['quantity']].sum()
```
