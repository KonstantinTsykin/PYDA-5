# Визуализация игр-бестселлеров в Европе на платформе PS4
Простой способ визуализировать какой продукт является наиболее ходовым на определенном рынке сбыта. Ниже добавляем аналогичный график для другой платформы и можем на основании визуализации сделать выводы о том, что есть как кросс-платформенные бестселлеры, так и игры эксклюзивно выходящие на определенной платформе, что часто является причиной выбора именно этой консоли.

```python
videogames = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df = videogames[videogames['Platform'] == 'PS4'].sort_values(by='EU_Sales', ascending = False).head(10)


plt.barh(df['Name'], df['EU_Sales'])
#top_sales_platform.info()
```

```python
videogames = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df = videogames[videogames['Platform'] == 'XOne'].sort_values(by='EU_Sales', ascending = False).head(10)


plt.barh(df['Name'], df['EU_Sales'])
```

# Таблица корреляций при помощи Seaborn
Демонстрирует корреляцию между столбцами таблицы. Позволяет выявить сильные взаимосвязи (например глобальные продажи "Global_Sales" очень сильно зависят от продаж на Северо-Американском рынке "NA_Sales") и приступить к дальнейшему анализу и проверке гипотез.

```python
sns.heatmap(videogames.corr(),annot=True,cmap='RdYlGn',linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```
