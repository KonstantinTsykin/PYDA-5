```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings_grouped = ratings.groupby('userId').count().reset_index()
ratings_filtered = ratings_grouped[ratings_grouped['rating'] >= 100]
film_fans_user_ids = ratings_filtered['userId'].tolist()
ratings[ratings['userId'].isin(film_fans_user_ids)]
#ratings['maxtime'] = ratings['timestamp'].max()
#ratings['mintime'] = ratings['timestamp'].min()
ratings['lifetime'] = ratings['timestamp'].max() - ratings['timestamp'].min()
ratings.groupby('userId').mean()
```
