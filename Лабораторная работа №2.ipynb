## Задание 6

```python
# Седаны дороже хэтчбеков
sedan_price=df[df['body-style']=='sedan']['price']
hatch_price=df[df['body-style']=='hatchback']['price']
```

```python
len(sedan_price), len(hatch_price)
```

```python
sedan_price = sedan_price.sample(65)
hatch_price = hatch_price.sample(65)
```

```python
statistics.variance(sedan_price), statistics.variance(hatch_price)
```

```python
sts.ttest_ind(sedan_price, hatch_price, equal_var=False)
```

```python
#полноприводные машины чаще дизельные, чем на бензине
fourwd_fuel = df[df['drive-wheels']=='4wd']['fuel-type']
wd_fuel = df[df['drive-wheels']!='4wd']['fuel-type']
```

```python
len(fourwd_fuel), len(wd_fuel)
```

```python
fourwd_fuel = sedan_price.sample(9)
wd_fuel = hatch_price.sample(9)
```

```python
statistics.variance(fourwd_fuel), statistics.variance(wd_fuel)
```

```python
sts.ttest_ind(fourwd_fuel, wd_fuel, equal_var=False)
```
